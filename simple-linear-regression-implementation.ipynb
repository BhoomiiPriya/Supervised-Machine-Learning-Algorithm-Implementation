```python
#Importing important libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

# Generate Sample Data

```python
np.random.seed(42)   #set random seed for reproducibility

X=2*np.random.rand(100,2)  #generate Y using a linear equation with some noise

Y=4+3*X+np.random.randn(100,1)  #generate Y using a linear equation with some noise
```

```python
print(X,Y)
```

```python
#Plot the data
plt.scatter(X,Y,color='red',alpha=0.6)
plt.xlabel("X (Independent Variable")
plt.ylabel("Y (Dependent Variable)")
plt.title("Scatter Plot of Data")
plt.show()
```

# Split the data into Training and Testing 

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Train the linear regression Model using Sk learn

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#get the learned parameter
print(f"Intercept: {model.intercept_[0]}")
print(f"Coefficient: {model.coef_[0][0]}")
```

# Make Prediction

```python
Y_pred=model.predict(X_test)
```

```python
print(Y_pred)
```

```python
print(Y_test)
```

# Evaluate the Model

```python
MSE= mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)
```

```python
print(f"Mean Squared Error: {MSE}")
print(f"R-Squared Score: {r2}")
```

# Plot the regression line

```python
plt.scatter(X_test,Y_test,color='blue',alpha=0.6,label="Actual Data")
plt.plot(X_test,Y_pred,color='red',linewidth=3,label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

# Implementing Linear Regression from Scratch

```python
#we will use Gradient Descent to find the best parameter beta0 and beta 1 means intercept and slope

#initialize parameters
beta_0=0
beta_1=0
learning_Rate=0.1
epochs=10
m=len(X)
```

# Gradient Descent Algorithm

```python
for i in range(epochs):
    Y_pred=beta_0 + beta_1 * X   #Predicted value
    error=Y_pred - Y             #error
    
    #compute Gradients
    gradient_beta_0=(2/m)*np.sum(error)
    gradient_beta_1=(2/m)* np.sum(error*X)
    
    #Update Parameter
    beta_0 = beta_0 -(learning_Rate*gradient_beta_0)
    beta_1 = beta_1 -(learning_Rate*gradient_beta_1)

#print final parameter
print(f"Intercept(beta_0):{beta_0}")
print(f"Slope (beta_1): {beta_1}")

    
```

# Plot Regression line

```python
plt.scatter(X,Y,color='blue',alpha=0.6)
plt.plot(X, beta_0 + beta_1*X,color='red',linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression(from scratch")
plt.show()
```
